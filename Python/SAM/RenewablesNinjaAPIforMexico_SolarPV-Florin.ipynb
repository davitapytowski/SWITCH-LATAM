```python
import requests
import pandas as pd
import json
import time
from io import StringIO
```

```python
token = '' #replace by your personal requested API token @ renewables.ninja
api_base = 'https://www.renewables.ninja/api/v1/'
```

```python
s = requests.session()
s.headers = {'Authorization':'Token ' + token}
s.headers
```

```python
countiesDF = pd.read_csv(r'MeshMexico.csv', header=-1,encoding = "ISO-8859-1")
#countiesDF = pd.read_csv(r'countiesLatLong.csv', header=-1,encoding = "ISO-8859-1")
countiesDF=countiesDF[1:]
countiesDF
```

```python
countiesDF[2].iloc[0] # just double checking the latitude
```

```python
countiesDF[3].iloc[0] # just double checking the longitude
```

```python
# For PV data

url = api_base + 'data/pv'
from tqdm import tqdm
for i,val in tqdm(enumerate(countiesDF[1]), desc="Counties in Mexico completed / # iterations"):
    time.sleep(73)

    args = {
    'lat': countiesDF[2].iloc[i],
    'lon': countiesDF[3].iloc[i],
    'date_from': '2014-01-01',
    'date_to': '2014-12-31',
    'capacity': 1000.0,
    'system_loss': 10,
    'tracking': 0,
    'tilt': (float(countiesDF[2].iloc[i])*0.76)+3.1, #http://www.alltrafficsolutions.com/SupportDocs/Optimum%20Orientation%20of%20Solar%20Panels.pdf
    'azim': 180, 
    'format': 'csv'
    }

    csv_string = s.get(url, params=args)
    csv_handle = StringIO(csv_string.text)
    valuesDF = pd.read_csv(csv_handle, index_col=0, parse_dates=True)
    filenameoutput='{}'.format(i+1)+".csv" 
    df=valuesDF.to_csv("/Users/sergiocastellanos/RenewableEnergyNinja/"+filenameoutput) # Replace folder
```
